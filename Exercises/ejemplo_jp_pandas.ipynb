```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = np.random.randn(100)
```

```python
ser = pd.Series(datos, pd.date_range('1/1/2020', periods=100))
```

```python
ser
```

```python
ser_acum = ser.cumsum()
```

```python
ser_acum.plot()
```
